# วิเคราะห์การถดถอยเชิงเส้นด้วยเทคนิคการเคลื่อนลงตามความชัน

<h3>การวิเคราะห์การถดถอย</h3>คือ การวิเคราะห์เพื่อหาความสัมพันธ์ระหว่างตัวแปรต้น(รู้ค่า)และตัวแปรตาม(ไม่รู้ค่า) เช่น การหาว่าหมาจะมีความต้องการปริมาณอาหาร(ตัวแปรตาม)เท่าใดเมื่อเวลา(ตัวแปรต้น)ผ่านไป x วัน ซึ่งจะได้สมการแสดงความสัมพันธ์ดังนี้

```python
from IPython.display import Math # to use math symbols
```

```python
Math(r' h=wx+b ')
```

เมื่อ w คือ ค่าน้ำหนักของตัวแปรต้น และ b คือ ค่าไบอัส 
ซึ่งการวิเคราะห์หาค่า w และ b นี้เรียก "การวิเคราะห์การถดถอยเชิงเส้น(Linear Regression Analysis)"

แต่ปัญหาทั่วไปอาจมีตัวแปรต้นหลายตัว ซึ่งจะได้สมการแสดงความสัมพันธ์ดังนี้

```python
Math(r' h=\sum_{j} w_{j}x_{j}+b=w_{1}x_{1}+ w_{2}x_{2}+...+ w_{n}x_{n}+b+b ')
```

โดย n เป็นจำนวนมิติ (จำนวนตัวแปรต้น) ของปัญหา ซึ่งในเบื้องต้นขอเริ่มจากปัญหาหนึ่งมิติดังในตัวอย่างที่ยกมานี้ก่อน จึงค่อยต่อยอดไปยังมิติที่สูงขึ้น

```python
import sys

import numpy as np

np.set_printoptions(threshold=sys.maxsize) # to show full np array

import matplotlib.pyplot as plt
x = np.arange(1,31,2)
z = 2.5+x*0.5+np.random.randn(15)*0.5
plt.gca(xlim=[0,31])
plt.xlabel(u'เวลา (วัน)',fontname='Tahoma', fontsize=15)
plt.ylabel(u'ปริมาณอาหาร (กก.)',fontname='Tahoma', fontsize=15)
plt.scatter(x,z)
plt.show()
```

จากสมการความสัมพันธ์ พบว่า w=0.5 และ b = 2.5 (ในที่นี้เราสุ่มเลขเพิ่มเข้ามาด้วย เพื่อเป็นตัวแทนของความไม่แน่นอนหรือสิ่งที่ควบคุมไม่ได้) ในการหา ml-model นั้นเราจะมองว่านี่คือชุดสร้าง traning data ที่เราจะทำไปใช้สอน test data เพื่อหา ml-model นั่นเอง ซึ่งในที่นี้เราจะใช้ ค่าความคลาดเคลื่อนในที่นี้ที่นิยมใช้ก็คือผลรวมความคลาดเคลื่อนกำลังสอง (sum of squared error, SSE) เพื่อหา w และ b ที่ทำให้ผลต่างระหว่างค่าของตัวแปรตามของจริงจาก test data (h : ค่าทำนาย) กับตัวแปรตามจาก traning data (z : ค่าจริง) มีน้อยที่สุด

note : w และ b ที่เราหาเป็นของ h

```python
Math(r'J=SSE=\sum_{i} (z_{(i)}-h_{(i)})^2 \\ h_{(i)}=wx_{(i)}+b')
```

โดย (i) เป็นดัชนีชี้ลำดับแถวของข้อมูล ที่เราจะนำความคลาดเคลื่อนที่คิดจากข้อมูลทั้งหมดมารวมกัน

ซึ่งเราจะ pre-code ได้ว่า
<p>h = w*x+b</p>
<p>sse = ((z-h)**2).sum()</p>

```python
from mpl_toolkits.mplot3d import Axes3D

plt.figure(figsize=[8,8])
ax = plt.axes([0,0,1,1],projection='3d',xlabel='b',ylabel='w',zlabel='SSE')
mb,mw = np.meshgrid(np.linspace(0,5,41),np.linspace(0,1,41))
sse = ((x*mw.ravel()[:,None]+mb.ravel()[:,None]-z)**2).sum(1).reshape(41,-1)
ax.plot_surface(mb,mw,sse,rstride=1,cstride=1,alpha=0.2)
plt.show()
```

ตำแหน่งของ w และ b ที่เรากำลังหา จะอยู่บริเวณจุดที่ SSE มีค่าต่ำที่สุด (แถวๆร่องตรงกลาง)

คราวนี้ลองเปลี่ยนเป็นวาดแผนภาพไล่สีแทน โดยสีแสดงถึงความสูง สีเหลืองคือค่าสูง สีดำคือค่าต่ำ จะเห็นว่าจุดที่ดำที่สุดก็คือแถว (2.5,0.5) ซึ่งเป็นคำตอบนั่นเอง

```python
import matplotlib as mpl
mb,mw = np.meshgrid(np.linspace(0,5,201),np.linspace(0,1,201))
sse = ((x*mw.ravel()[:,None]+mb.ravel()[:,None]-z)**2).sum(1).reshape(201,-1)
plt.gca(xlim=[0,5],ylim=[0,1])
plt.pcolormesh(mb,mw,sse,norm=mpl.colors.LogNorm(),cmap='gnuplot')
plt.colorbar(pad=0.01)
plt.show()
```

หากดูจากรูปเราก็คงพอเข้าใจคร่าวๆได้ว่าคำตอบที่ควรจะเป็นตรงไหน แต่ทีนี้เราจะเขียนโปรแกรมยังไงให้มันหาคำตอบนั้นเจอได้เอง นั่นคือโจทย์ในตอนนี้

วิธีการหาคำตอบนั้นอาจมีอยู่หลายวิธี แต่ในที่นี้จะแนะนำวิธีที่เรียกว่าการเคลื่อนลงตามความชัน (Gradient Descent, GD) ซึ่งนิยมใช้มาก

<h3>Gradient Descent</h3>

สมมุติว่าเราทำเข็มอันหนึ่งตกลงไปในมหาสมุทร แต่เรารู้ว่าของมีน้ำหนักย่อมตกสู่ที่ต่ำ ดังนั้นถ้าจะหาเข็มให้เจอเราก็ต้องหาพื้นใต้สมุทรส่วนที่อยู่ต่ำที่สุด

เราอาจเริ่มจากดำน้ำลงจุดไหนสักจุดก่อนแล้วดูว่าจุดนั้นมีความลาดเอียงแค่ไหน ถ้าจุดที่เราดำน้ำลงมาเจอพื้นนั้นไม่ใช่จุดต่ำที่สุดเราก็จะต้องรู้สึกได้ถึงความลาดเอียงของพื้น แล้วก็จะสามารถคิดได้ว่าหากอยากไปที่ที่ต่ำลงไปก็ต้องเคลื่อนที่ไปตามความลาดเอียงนั้น

นั่นคือแนวคิดของเทคนิคการเคลื่อนลงการตามความชันก็คือหาค่าความชันของตำแหน่งที่อยู่ แล้วใช้มันเป็นตัวกำหนดทิศทางว่าจะปรับเปลี่ยนตำแหน่งไปทางไหนต่อ

"ตำแหน่ง" ในที่นี้หมายถึงค่า w และ b ในขณะที่ "ความสูง" คือค่า SSE

"ความชัน" ในที่นี้คืออนุพันธ์ของ SSE โดยคิดเป็นอนุพันธ์ย่อยเทียบกับ w และ b นั่นคือ

```python
Math(r' \frac{\partial \mathbf{J}}{\partial \mathbf{w}}=2\sum_{i} (z_{(i)}-h_{(i)})x_{(i)} ')
```

```python
Math(r' \frac{\partial \mathbf{J}}{\partial \mathbf{b}}=2\sum_{i} (z_{(i)}-h_{(i)}) ')
```

```python
Math(r' \Delta w = -\eta\frac{\partial \mathbf{J}}{\partial \mathbf{w}}=2\eta\sum_{i} (h_{(i)}-z_{(i)})x_{(i)} ')
```

```python
Math(r' \Delta b = -\eta\frac{\partial \mathbf{J}}{\partial \mathbf{b}}=2\eta\sum_{i} (h_{(i)}-z_{(i)}) ')
```

โดย <b>η</b> คือสิ่งที่เรียกว่า "อัตราการเรียนรู้" เป็นค่าที่กำหนดว่าในแต่ละรอบที่มีการปรับค่าน้ำหนักและไบแอสจะปรับมากแค่ไหน

ค่านี้มีความสำคัญมากต้องกำหนดให้เหมาะกับปัญหา ถ้ามากเกินไปน้ำหนักจะถูกปรับมากเกินจนไม่อาจลู่เข้าสู่คำตอบได้ แต่ถ้าน้อยไปการปรับน้ำหนักก็จะแทบไม่คืบหน้า

```python
eta = 0.0002 # อัตราการเรียนรู้
n_thamsam = 10000 # จำนวนครั้งที่ทำซ้ำเพื่อเรียนรู้
w,b = 0,0 # น้ำหนักและไบแอสเริ่มต้น
wi = [w] # ลิสต์บันทึกค่าน้ำหนักและไบแอส
bi = [b]
h = w*x+b # คำนวณคำตอบโดยใช้ w และ b ตอนแรก
for i in range(n_thamsam):
    w += 2*((z-h)*x).sum()*eta # ปรับค่าน้ำหนักและไบแอส
    b += 2*(z-h).sum()*eta
    wi += [w] #  บันทึกค่าน้ำหนักและไบแอสใหม่
    bi += [b]
    h = w*x+b # คำนวณคำตอบโดยใช้ค่า w และ b ใหม่
```

```python
plt.gca(xlim=[0,31])
plt.xlabel(u'เวลา (วัน)',fontname='Tahoma')
plt.ylabel(u'ปริมาณอาหาร (กก.)',fontname='Tahoma')
plt.scatter(x,z)
xsen = np.array([0,31])
ysen = xsen*w+b
plt.plot(xsen,ysen,'b')
plt.show()
```

```python
# Plot Line
bi = np.array(bi)
wi = np.array(wi)
plt.figure(figsize=[8,8])
ax = plt.axes([0,0,1,1],projection='3d',xlabel='b',ylabel='w',zlabel='SSE')
ssei = ((x*wi[:,None]+bi[:,None]-z)**2).sum(1)
ax.plot(bi,wi,ssei,'bo-')
# Plot Space
mb,mw = np.meshgrid(np.linspace(0,3,201),np.linspace(0,1.2,201))
sse = ((x*mw.ravel()[:,None]+mb.ravel()[:,None]-z)**2).sum(1).reshape(201,-1)
ax.plot_surface(mb,mw,sse,rstride=5,cstride=5,alpha=0.2,color='b',edgecolor='k')
```

```python
plt.figure(figsize=[10,5])
plt.pcolormesh(mb,mw,sse,norm=mpl.colors.LogNorm(),cmap='gnuplot')
plt.colorbar(pad=0.01)
plt.plot(bi,wi,'bo-')
plt.show()
```

เราอาจกำหนดเกณฑ์ในการให้หยุดโดยให้หยุดเมื่อค่า SSE เปลี่ยนแปลงน้อยมากก็ได้ แต่ในที่นี้จะใช้การเปลี่ยนแปลงค่า w และ b เป็นตัวกำหนด 

เช่น ลองกำหนดให้หยุดเมื่อทั้งค่า w และ b เปลี่ยนแปลงน้อยกว่า 0.0000001

```python
eta = 0.0002
n_thamsam = 100000
d_yut = 1e-7 # ค่าความเปลี่ยนแปลงน้ำหนักและไบแอสสูงสุดที่จะให้หยุดได้
w,b = 0,0
h = w*x+b
for i in range(n_thamsam):
    dw = 2*((z-h)*x).sum()*eta
    db = 2*(z-h).sum()*eta
    w += dw
    b += db
    h = w*x+b
    if(np.abs(dw)and np.abs(db)<d_yut):
        break # หยุดเมื่อทั้ง dw และ db ต่ำกว่า d_yut

print('ทำซ้ำไป %d ครั้ง dw=%.3e, db=%.3e'%(i,dw,db))
# ทำซ้ำไป 7064 ครั้ง dw=-5.006e-09, db=9.993e-08
```
